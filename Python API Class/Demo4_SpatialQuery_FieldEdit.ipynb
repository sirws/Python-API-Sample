```python
import arcgis
from getpass import getpass
```

```python
gis = arcgis.gis.GIS("https://www.arcgis.com", "ScottMoorePNW", getpass())
```

```python
county_lyr = arcgis.features.FeatureLayer("https://data.wsdot.wa.gov/arcgis/rest/services/Shared/CountyBoundaries/MapServer/0")
```

```python
county_fset = county_lyr.query(where="JURNM='Thurston County'", return_geometry=True)
```

```python
county_fset.features
```

```python
county_geom = county_fset.features[0].geometry
```

```python
county_geom
```

```python
county_geom["spatialReference"] = county_fset.spatial_reference
```

```python
county_geom
```

```python
items = gis.content.search(query="title:WildlifeObservations_ScottMoorePNW", item_type="Feature Layer")
```

```python
if len(items) > 0:
    item = items[0]
else:
    print("Your search returned no records.  Game over.")
```

```python
lyr = item.layers[0]
```

```python
fset = lyr.query(geometry_filter=arcgis.geometry.filters.intersects(county_geom), out_sr=lyr.properties.extent.spatialReference)
```

```python
features = fset.features
```

```python
features
```

```python
updates_list = []
for feat in features:
    feat.set_value("County", "Thurston")
    updates_list.append(feat)
```

```python
updates_list
```

```python
lyr.edit_features(updates=updates_list)
```
