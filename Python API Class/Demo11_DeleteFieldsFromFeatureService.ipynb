```python
import arcgis
from getpass import getpass
```

```python
gis = arcgis.gis.GIS("https://www.arcgis.com", "YOURUSERNAME", getpass())
```

```python
item = gis.content.search(query="title:WildlifeObservations_YOURUSERNAME", item_type="Feature Layer")[0]
```

```python
lyr = item.layers[0]
```

```python
lyr.properties
```

```python
del_flds = []
lyr_props = dict(lyr.properties)
for fld_dict in lyr_props["fields"]:
    if fld_dict["name"].lower().startswith("sensitive"):
        del_flds.append(fld_dict)
```

```python
del_flds
```

```python
del_dict = {"fields": del_flds}
```

```python
lyrmgr = lyr.manager
```

```python
lyrmgr._hydrate()
```

```python
lyrmgr.delete_from_definition(del_dict)
```
